```python

# imports needed and set up logging
import gzip
import gensim 
import logging
import json
from gensim.models import word2vec as w2v

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python

fileArr = ['./recipes_raw_nosource_ar.json','./recipes_raw_nosource_epi.json','./recipes_raw_nosource_fn.json']
def readFile(fileData):
  dataDic = json.load(fileData)

  instArr = []
  for each in dataDic:
    value = dataDic[each]
    instruction = value.get('instructions')
    instArr.append(instruction)
  return instArr

for eachFile in fileArr:
  fileData = open(eachFile)
  documents = readFile(fileData)
  

    





```

```python

cleaned = [x for x in documents if x != None]
cleaned
```

```python
def refineData():
  updatedArr = []
  for i in cleaned:
      updatedArr.append(gensim.utils.simple_preprocess(i))
  return updatedArr

generated = refineData()
```

```python
generated
```

```python
model = w2v.Word2Vec(min_count=1)
model.build_vocab(generated)
model.train(generated,total_examples=len(generated),epochs=10)
```

```python
print(model.wv.most_similar(positive=['pasta','taco','rice','lettuce']))
```
